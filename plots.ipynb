```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

files = {
    "4 Batch Size": "/content/log_summary_4BS.xlsx",
    "8 Batch Size": "/content/log_summary_8BS.xlsx",
    "16 Batch Size": "/content/log_summary_16BS.xlsx",
    "32 Batch Size": "/content/log_summary_32BS.xlsx"
}
```

```python
# Time
plt.figure(figsize=(12, 5))
for label, file_path in files.items():
    df = pd.read_excel(file_path)
    time_col = next((col for col in df.columns if "minibatch" in col.lower() and "time" in col.lower()), None)
    if time_col:
        plt.plot(df[time_col].values, label=label)


plt.title("Minibatch Time Comparison Across Batch Sizes")
plt.xlabel("Step Index")
plt.ylabel("Minibatch Time")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Accuracy
plt.figure(figsize=(12, 5))
for label, file_path in files.items():
    df = pd.read_excel(file_path)

    acc_col = next((col for col in df.columns if "accuracy" in col.lower()), None)
    if acc_col:
        plt.plot(df[acc_col].values, label=label)
plt.title("Accuracy Comparison Across Batch Sizes")
plt.xlabel("Step Index")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
